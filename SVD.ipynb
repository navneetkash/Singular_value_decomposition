```python
%matplotlib inline
from skimage import data
import matplotlib.pyplot as plt
camera = data.camera()
print(type(camera),camera.dtype,camera.shape)
plt.imshow(camera,cmap='gray', interpolation='nearest');
```

```python
print(camera)
```

```python
import numpy.linalg as la
import numpy as np
```

```python
U, s, VT = la.svd(camera)
print(U.shape)
print(s.shape)
print(VT.shape)

x = np.arange(s.shape[0])
plt.scatter(x,s)
```

```python
from numpy import diag
from numpy import dot
from numpy import zeros
```

```python
sigma = zeros((camera.shape[0], camera.shape[1]))
for it in range(s.shape[0]):
    if it > 50:
        s[it] = 0
print(s)
```

```python
sigma[:camera.shape[0], :camera.shape[0]] = diag(s)
B = U.dot(sigma.dot(VT))
plt.imshow(B, cmap='gray', interpolation='nearest')
```
